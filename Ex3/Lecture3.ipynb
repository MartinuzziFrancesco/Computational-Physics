# Inverse transformation method
Generating random numbers with a given distribution, in this case exponential. the first graph is the result of the binning done by the program, the second is done with python, using the raw data.

```python
exp = pd.read_csv('expdev.dat', header=None, sep='\s+')
raw = pd.read_csv('rawdata.dat', header=None, sep='\s+')
```

```python
x = exp[0]
y = exp[1]

fig = plt.figure(figsize=(20, 10))
ax = fig.add_subplot(111)
ax.plot( x, y)
ax.set_yscale("log", nonposy='clip')
ax.grid(False)
```

```python
exp[1].plot(figsize=(20, 10))
```
