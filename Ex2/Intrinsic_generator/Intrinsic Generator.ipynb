```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv('random1.dat', header=None, sep='\s+')
df1.columns = ['i', 'rndm']
df1.head()
```

```python
df1['rndm'].hist(bins=100, figsize=(20, 10))
```

```python
df1.plot(x='i', y='rndm', kind='scatter', figsize=(20, 10))
```

```python
df2 = pd.read_csv('random2.dat', header=None, sep='\s+')
df2.columns = ['i', 'rndm']
df2.head()
```

```python
df2['rndm'].hist(bins=100, figsize=(20, 10))
```

```python
df2.plot(x='i', y='rndm', kind='scatter', figsize=(20, 10))
```

```python
comp1 = df1.iloc[::2]
comp2 = df1.iloc[1::2]

x=comp1['rndm']
y=comp2['rndm']

fig = plt.figure(figsize=(20, 10))
ax = fig.add_subplot(111)
ax.scatter(x, y)
ax.grid(True)
```
